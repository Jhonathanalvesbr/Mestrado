```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib
import math
import os
import json
```

```python
# Create an empty directed network
DG = nx.DiGraph()

# Add nodes
a = [1, 2, 3, 4, 5, 6]
DG.add_nodes_from(a)

# Add edges
b = [(1, 2),(1, 4), (1, 5), (2, 4), (2, 6), (3, 4), (5,6)]
DG.add_edges_from(b)
size = dict(DG.degree)

# Draw the directed network #lightgray
cmap = matplotlib.colors.ListedColormap(['dodgerblue', 'darkorange', 'lightgray'])
maximo = -math.inf
valores = []
for v in size.values():
    v = (v * 1000)
    valores.append(v)
    if(maximo < v):
        maximo = v
        
node_color = []

lista = list(size.values())
soma = sum(lista)
valores = []
for k in lista:
    valores.append(int((k*100)/soma))

node_color = []
for k in valores:
    if(k < 33):
        node_color.append(0)
    elif(k >= 33 and k <= 66):
        node_color.append(1)
    elif(k > 66):
        node_color.append(2)
nx.draw(DG, node_size=[1 for v in a])
plt.show()
```

```python
import os
import json
from datetime import datetime
from datetime import date
import time
pasta = "C:\\Users\\admin\\Desktop\\Nova pasta"

class Noti():
    def __init__(self):
        self.noticia = {}
        self.tweets = {}
        self.retweets = {}

class Tweet():
    def __init__(self, tweet, retweet):
        self.tweet = tweet
        self.retweet = retweet
        self.nRetweet = 1
        self.nSeguidores = 1

class Noticia():
    def __init__(self):
        self.noticia = ""
        self.tweet = []

noti = Noti()

for diretorio, subpastas, arquivos in os.walk(pasta):
    for arquivo in arquivos:
        caminho = os.path.join(os.path.realpath(diretorio), arquivo)
        if(diretorio.find("retweets") > -1):
            noti.retweets[arquivo] = caminho
        elif(diretorio.find("tweets") > -1):
            noti.tweets[arquivo] = caminho
        else:
            noti.noticia[arquivo] = caminho

noticia = Noticia()

for k in noti.tweets:
    if((k in noti.retweets) == True):
        f = open(noti.tweets[k])
        twee = json.load(f)
        f = open(noti.retweets[k])
        retwee = json.load(f)
        tweet = Tweet(twee,retwee)
        noticia.tweet.append(tweet)
    else:
        f = open(noti.tweets[k])
        twee = json.load(f)
        tweet = Tweet(twee,None)
        noticia.tweet.append(tweet)
relacionamentos = []
nodes = []
criado = []
contarSeguidores = 0
totalN = []
nPerfis = []
for k in noticia.tweet:
    if((k.tweet['user']['id'] in nPerfis) == False):
        nPerfis.append(k.tweet['user']['id'])
        
    nodes.append(k.tweet['id'])
    k.nSeguidores = int(k.tweet['user']['followers_count'])
    if((k.tweet['user']['id'] in totalN) == False):
        totalN.append(k.tweet['user']['id'])
        contarSeguidores += int(k.tweet['user']['followers_count'])
    new_datetime = datetime.strftime(datetime.strptime(k.tweet['created_at'],'%a %b %d %H:%M:%S +0000 %Y'), '%H:%M:%S %d/%m/%Y')
    criado.append(new_datetime)
    if(k.retweet != None):
        for n in k.retweet["retweets"]:
            k.nRetweet += 1
            contarSeguidores += int(n['user']['followers_count'])
            if((n['user']['id'] in totalN) == False):
                totalN.append(n['user']['id'])
                contarSeguidores += int(n['user']['followers_count'])
            relacionamentos.append((k.tweet['id'], n['id']))
            new_datetime = datetime.strftime(datetime.strptime(n['created_at'],'%a %b %d %H:%M:%S +0000 %Y'), '%H:%M:%S %d/%m/%Y')
            criado.append(new_datetime)
print("14:03:24 11-12-2016\n\n")


sorted_list = sorted(criado, key=lambda t: datetime.strptime(t, '%H:%M:%S %d/%m/%Y'))

for k in sorted_list:
    print(k)

d0 = date(int(sorted_list[0].split(" ")[1].split("/")[2]), int(sorted_list[0].split(" ")[1].split("/")[1]), int(sorted_list[0].split(" ")[1].split("/")[0]))
d1 = date(int(sorted_list[-1].split(" ")[1].split("/")[2]), int(sorted_list[-1].split(" ")[1].split("/")[1]), int(sorted_list[-1].split(" ")[1].split("/")[0]))
delta = d1 - d0
print("Dias: " + str(delta.days))
print("Seguidores total: " + str(contarSeguidores))
print("Pefils: " + str(len(nPerfis)))
```
